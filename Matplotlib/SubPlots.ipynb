```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots()
```

```python
axes
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
type(axes)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
plt.tight_layout();
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,dpi=200)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
fig.subplots_adjust(wspace=0.5,hspace=0.5)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,dpi=200)
axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].set_ylabel("Y label 1.0")
axes[1][0].set_xlabel("X label 1.0")
axes[1][0].set_title("Title 1.0")
fig.subplots_adjust(wspace=0.5,hspace=0.5)
fig.suptitle("FIGURE LEVEL TITLE")
```

```python
fig.savefig("new_subplots.png")
```
