```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

# .legend()

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, label="X vs X")
ax.plot(x,x**2, label="X vs X^2")
ax.legend(loc=(1.01,0.5));
```

# Colors and styles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color = "#34eb9b", label="X vs X", lw=5)
ax.plot(x,x+1, color="purple", label= "X vs (X+1)", lw=2)
ax.legend();
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color = "#34eb9b", label="X vs X")
ax.plot(x,x+1, color="purple", label= "X vs (X+1)")
ax.legend();
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color = "#34eb9b", label="X vs X", ls="--")
ax.plot(x,x+1, color="purple", label= "X vs (X+1)", ls="-.")
ax.legend();
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x, color = "#34eb9b", label="X vs X", marker="+", ls="--")
ax.plot(x,x+1, color="purple", label= "X vs (X+1)", marker=".", ls="-.")
ax.legend();
```
