```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a ** 4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2 * x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
axes.plot(a,b);
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Axes1")

# SMALL AXES
axes2 = fig.add_axes([0.15,0.3,0.5,0.5])
axes2.plot(x,y,"r");
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("Axes2")
```

```python
fig = plt.figure(figsize=(6,4),dpi=200)
axes3 = fig.add_axes([0,0,1,1])
axes3.plot(a,b)
fig.savefig("new_figure.png", bbox_inches="tight");
```
