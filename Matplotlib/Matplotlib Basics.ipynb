```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2 * x
```

```python
plt.plot(x,y)
plt.title("String Title")
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("String Title")
```

```python
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.xlim(0,6)
plt.ylim(0,10)
plt.title("String Title");
```
