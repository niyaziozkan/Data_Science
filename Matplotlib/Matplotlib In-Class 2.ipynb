```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary1 = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
salary2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary1, "g", label="Salary - 1")
plt.plot(age, salary2, "y", label="Salary - 2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age by Salary")
plt.legend();
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary1, "g", label="Salary - 1")
plt.legend()
plt.subplot(1,2,2)
plt.plot(age, salary2, "y", label="Salary - 2")
plt.legend()
plt.tight_layout();
```

```python
fig,ax=plt.subplots()
ax.plot(age, salary1, "r", label="Salary - 1")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Age by Salary");
```

```python
fig,ax=plt.subplots(2,2)
ax[0,0].plot(age,salary1,"r")
ax[0,0].set_title("Age by Salary - 1")
ax[0,1].plot(age,salary2,"g")
ax[0,1].set_title("Age by Salary - 2")
plt.tight_layout();
```

```python
fig,ax=plt.subplots(dpi=100)
ax.plot(age, salary1, "r", label="Salary")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_xlim([30,40])
ax.set_ylim([45000,70000])
ax.axvline(34, ls="--")
ax.axhline(60000, ls="-.", c="y")
ax.set_title("Age by Salary");
```

```python
country = ["UK", "USA", "FRA", "GER", "NOR"]
pci = [40000, 50000, 38000, 55000, 80000]
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
langs = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
```

```python
fig,ax = plt.subplots(dpi=150)
ax.pie(slices, labels=langs, autopct="%1.2f%%", explode=(0,0,0.1,0,0), shadow=True)
ax.axis("equal");
```
