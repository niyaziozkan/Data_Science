```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
age
```

```python
salary
```

# Matplotlib Functional Method

```python
plt.plot(age, salary);
```

```python
plt.plot(age, salary)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary by Age");
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
salary_2
```

```python
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary by Age");
```

```python
plt.plot(age, salary, label = "Salary 1")
plt.plot(age, salary_2, label = "Salary 2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary by Age")
plt.legend();
```

```python
plt.figure(figsize=(8,8))
plt.plot(age, salary, label = "Salary 1")
plt.plot(age, salary_2, label = "Salary 2")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Salary by Age")
plt.legend();
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary, "b")
plt.subplot(1,2,2)
plt.plot(age, salary_2, "y")
plt.tight_layout();
```

```python
plt.subplot(2,1,1)
plt.plot(age, salary, "b")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "y")
plt.tight_layout();
```

```python
plt.figure(figsize = (8,4))
plt.subplot(1,2,1)
plt.plot(age, salary, "b")
plt.subplot(1,2,2)
plt.plot(age, salary_2, "y")
plt.tight_layout();
```

```python
fig = plt.figure()
ax = fig.add_axes([0.0,0.0,0.8,0.8]) #left, bottom, width, height (range 0 to 1)
ax.plot(age,salary, "b")
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age");
```

# Matplotlib Object Oriented Method

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0,0.0,0.8,0.8]) #left, bottom, width, height (range 0 to 1)
ax2 = fig.add_axes([0.95,0,0.4,0.4]) #left, bottom, width, height (range 0 to 1)
ax1.plot(age,salary, "b")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")

ax2.plot(age,salary_2, "y")
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary")
ax2.set_title("Salary by Age");
```
