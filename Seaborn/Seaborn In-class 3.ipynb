# Grids

## 1. catplot()

```python
sns.catplot(x="sex", y="total_bill", data=tips, kind="box");
```

## 2. PairGrid()

```python
g = sns.PairGrid(tips)
g.map(plt.scatter);
```

```python
g = sns.PairGrid(tips)
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, color="r");
```

```python
g = sns.PairGrid(mpg)
g.map_diag(sns.rugplot, color="r")
g.map_upper(sns.scatterplot, color="g")
g.map_lower(sns.histplot, color="b");
```

```python
## FacetGrid()
```

```python
g = sns.FacetGrid(tips, col="time")
g.map(sns.kdeplot, "total_bill")
```

```python
g = sns.FacetGrid(tips, row="smoker")
g.map(sns.kdeplot, "total_bill")
```

```python
sns.boxplot(x="vehicleType", y="price", data=df)
```

```python
plt.figure(figsize=(12,7))
sns.pointplot(x="vehicleType", y="price", data=df, order=df.groupby("vehicleType")["price"].mean().sort_values(ascending=False).index);
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x="vehicleType", data=df, order=df["vehicleType"].value_counts().sort_values(ascending=False).index);
```
