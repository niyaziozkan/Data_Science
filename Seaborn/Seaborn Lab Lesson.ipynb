```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

## scatterplot()

```python
sns.scatterplot(y="tip", x="total_bill", data=tips, color="green", size="size", marker="+", hue="sex")
```

```python
fig, ax = plt.subplots()
tips.plot(kind="scatter", x="total_bill", y="tip",c="size", ax=ax);
```

## Distribution Plots

### Rugplot

```python
sns.rugplot(x="total_bill",y="tip", data=tips, height=0.2)
```

### Histogram

```python
sns.histplot(x="tip", data=tips, bins=8, kde=True, ls="--");
```

### Displot

```python
sns.displot(x="tip", data=tips,rug=True)
```

### Kdeplot

```python
sns.kdeplot(x="tip", data=tips)
```

## Categorical Plots

### Count Plot

```python
plt.figure()
sns.countplot(x="sex",data=tips, hue="smoker", palette="magma");
```

## Grid Plots

### PairGrid

```python
g = sns.PairGrid(tips, hue="sex")
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot);
```
