```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

## countplot()

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.countplot(x="division", data=df);
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.countplot(x="level of education", data=df, hue="division", palette="tab20");
```

```python
df["level of education"].value_counts()
```

## barplot()

```python
plt.figure(figsize=(10,4),dpi=100)
sns.barplot(data=df, x="level of education", y="salary", estimator=np.mean, ci="sd", hue="division")
plt.legend(bbox_to_anchor=(1.05,1));
```

# Distribution within Categories

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

## boxplot()

```python
sns.boxplot(x="math score", data=df);
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.boxplot(x="math score",y="test preparation course", data=df);
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.boxplot(x="reading score",y="test preparation course", data=df);
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.boxplot(x="reading score",y="parental level of education", data=df);
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.boxplot(x="reading score",y="parental level of education",hue="test preparation course", data=df)
plt.legend(bbox_to_anchor=(1.05,1));
```

## violinplot()

```python
plt.figure(figsize=(10,8),dpi=100)
sns.violinplot(x="reading score",y="parental level of education",data=df);
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.violinplot(x="reading score",y="parental level of education",hue="test preparation course",data=df)
plt.legend(bbox_to_anchor=(1.23,1));
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.violinplot(x="reading score",y="parental level of education",hue="test preparation course",data=df, split=True)
plt.legend(bbox_to_anchor=(1.23,1));
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.violinplot(x="reading score",y="parental level of education",data=df, inner="quartile");
```

```python
plt.figure(figsize=(10,8),dpi=100)
sns.violinplot(x="reading score",y="parental level of education",data=df, inner="sticks");
```

## swarmplot()

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(x="math score", data=df, size=2)
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(x="math score",y="gender", data=df, size=2)
```

```python
plt.figure(figsize=(8,4),dpi=200)
sns.swarmplot(x="math score",y="gender",hue="test preparation course",dodge=True, data=df, size=2)
plt.legend(bbox_to_anchor=(1.23,1));
```
