```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x="salary",y="sales",data=df,hue="level of education", palette="Dark2");
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x="salary",y="sales",data=df,hue="salary");
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x="salary",y="sales",data=df,size="work experience");
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x="salary",y="sales",data=df,s=200, alpha=0.2);
```

```python
plt.figure(figsize=(12,4),dpi=200)
sns.scatterplot(x="salary",y="sales",data=df,s=200, style="level of education", hue="level of education")
plt.savefig("seaborn scatter plot.jpg");
```
