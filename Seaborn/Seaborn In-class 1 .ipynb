# Distribution Plots

## 1. rugplot()

```python
mpg.head()
```

```python
sns.rugplot(x="weight", height=0.9, data=mpg);
```

## 2. displot() and histplot()

```python
sns.displot(x="weight", data=mpg);
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.histplot(data=mpg,x='weight', kde=True, ax=ax[0])
sns.rugplot(x='weight',data=mpg, ax=ax[1]);
```

```python
sns.set(style="darkgrid")
sns.histplot(data=mpg,x='weight', kde=False);
```

```python
planets.head()
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
sns.histplot(data=planets,x='mass', ax=ax[0])
sns.rugplot(x='distance',data=planets, ax=ax[1]);
```

```python
penguins.head()
```

```python
sns.displot(data=penguins, x="body_mass_g",rug=True, kde=True);
```

## 3. kdeplot()

```python
sns.kdeplot(data=penguins, x="body_mass_g", shade=True, color="red");
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips);
```

```python
sns.kdeplot(x="total_bill", y="tip", data=tips, shade=True, color="red", alpha=0.8);
```

```python
fig,ax = plt.subplots(1,2, figsize=(10,5))
sns.scatterplot(x='total_bill', y='tip', data=tips, ax=ax[0])
sns.kdeplot(x='total_bill', y='tip', data=tips, ax=ax[1], shade=True, color='red');
```
