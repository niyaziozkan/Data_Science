```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

# COMPARISON PLOTS

## jointplot()

```python
sns.jointplot(data=df, x="math score", y="reading score");
```

```python
sns.jointplot(data=df, x="math score", y="reading score", kind="hex", color="green", alpha=0.7);
```

```python
sns.jointplot(data=df, x="math score", y="reading score", kind="hist");
```

```python
sns.jointplot(data=df, x="math score", y="reading score", kind="kde", shade=True);
```

```python
sns.jointplot(data=df, x="math score", y="reading score", hue="gender");
```

## pairplot()

```python
sns.pairplot(df);
```

```python
sns.pairplot(df, hue="gender");
```

```python
sns.pairplot(df, hue="gender", diag_kind="hist");
```

```python
sns.pairplot(df, hue="gender", corner=True);
```

# GRID PLOTS

```python
df.head()
```

## catplot()

```python
sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch", row="test preparation course");
```

## PairGrid()

```python
g = sns.PairGrid(df, hue="gender")
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
g.add_legend();
```

## MATRIX PLOTS

## heatmap()

```python
df = pd.read_csv("country_table.csv")
```

```python
df
```

```python
df = df.set_index("Countries")
```

```python
df
```

```python
sns.heatmap(df);
```

```python
plt.figure(dpi=100)
sns.heatmap(df.drop("Life expectancy", axis=1), lw=0.5, annot=True, cmap="cividis", center=15)
```

## clustermap()

```python
plt.figure(dpi=100)
sns.clustermap(df.drop("Life expectancy", axis=1), lw=0.5, annot=True, cmap="cividis", col_cluster=False)
```
