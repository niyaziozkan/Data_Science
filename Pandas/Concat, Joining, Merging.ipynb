# Merging


```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right)
```

```python
pd.merge(left,right, how="inner", on = "key")
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K4', 'K5'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how="inner")
```

```python
pd.merge(left,right,how="outer", on="key")
```

```python
pd.merge(left,right,how="left", on="key")
```

```python
pd.merge(left,right,how="right", on="key")
```

```python
df11 = pd.DataFrame({'lkey': ['x', 'y', 'z', 'x', 'z'],
                    'lvalue': [2, 3, 5, 7, 0]})
df12 = pd.DataFrame({'rkey': ['a', 'x', 'z', 'b'],
                    'rvalue': [7, 8, 9, 10]})
```

```python
df11
```

```python
df12
```

```python
pd.merge(df11,df12,left_on="lkey", right_on="rkey")
```

```python
pd.merge(df11,df12,left_on="lkey", right_on="rkey", how="outer")
```

```python
pd.merge(df11,df12,left_on="lkey", right_on="rkey", how="left")
```

```python
pd.merge(df11,df12,left_on="lkey", right_on="rkey", how="right")
```

# Joining

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index = ['K0', 'K1', 'K2'])
right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index = ['K0', 'K2', 'K3'])
```

```python
left
```

```python
right
```

```python
left.join(right)
```

```python
left.join(right, how="inner")
```

```python
left.join(right, how="outer")
```

```python
df = pd.DataFrame({'key': ['K0', 'K2', 'K3', 'K4', 'K5', 'K6'],
                   'X': ['X0', 'X2', 'X3', 'X4', 'X5', 'X6']})
df
```

```python
other = pd.DataFrame({'key': ['K0', 'K2', 'K3'],
                      'Y': ['Y0', 'Y2', 'Y3']})
other
```

```python
df.join(other, lsuffix="_df", rsuffix="_other")
```

```python
df.set_index("key").join(other.set_index("key"))
```

```python
df.join(other.set_index("key"), on="key")
```
